### Applying K-Means for clustering data and assigning labels

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
# creating dummies for object features
dummy_df = pd.get_dummies(df[['ot','home_team','away_team']])
dummy_df
```

```python
# adding dummies_df to original df
df = pd.concat([df,dummy_df],axis=1)
df
```

```python
model_df = df.iloc[:,3:]
model_df
```

```python
data_cols = list(model_df.columns)
```

```python
# Applying MinMax scaler for scaling the data.
from sklearn.preprocessing import MinMaxScaler
mms_scale=MinMaxScaler(feature_range=(0,2))
m_data=mms_scale.fit_transform(model_df)
np.set_printoptions(precision=3)
print(m_data)
```

```python
model_df = m_data
```

### Automatic feature selection using PCA

```python
# Doing  PCA giving number of Components(dimensions). checking by reducing to 2 components
pca_comp2=PCA(n_components=2)
x_pca_comp=pca_comp2.fit_transform(model_df)
x_pca_comp
```

```python
model_df = x_pca_comp
```

```python
# initializing and clustering the data
k_model = KMeans(n_clusters=4, random_state=21)
kmeans = k_model.fit(model_df)
```

```python
# checking the cluster centers
kmeans.cluster_centers_
```

```python
# checking the spread of clusters
kmeans.inertia_
```

```python
# labelling all the rows
labels = kmeans.predict(model_df)
labels
```

```python
# calculate spread for a range of number of cluster
spreads = []
for i in range(1, 15):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(model_df)
    spreads.append(kmeans.inertia_)

# plot
plt.figure(figsize=(7, 5), dpi=80)
plt.plot(range(1, 15), spreads, marker='o')
plt.xlabel('Number of centroids',fontsize=12)
plt.xticks(list(range(1,15)),fontsize=12)
plt.ylabel('Spread',fontsize=12)
plt.yticks(fontsize=12)
plt.title("Elbow method for number of centroids vs Inertia",fontsize=12)
plt.show()
```

```python
# Applying 3 cluster model to the data and labelling all the rows
# initializing and clustering the data
k_model = KMeans(n_clusters=3, random_state=25)
kmeans = k_model.fit(model_df)
```

```python
kmeans.inertia_
```

```python
# labelling all the rows
labels = kmeans.predict(model_df)
labels
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
# centroids graph for K-Means
colors = ['blue', 'green', 'orange']
markers = ['o', 'v', 's']
plt.figure(figsize=(7, 5), dpi=80)
for i, l in enumerate(kmeans.labels_):
    plt.plot(model_df[i][0], model_df[i][1], color=colors[l], marker=markers[l],ls='None')

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], marker="x", color='r', s=1000)
plt.title("Centroids of Dimensions Reduced Data",fontsize=12)
plt.xticks([-2,-1,0,1,2],fontsize = 12)
plt.yticks(fontsize = 12)
plt.show()
```

```python
plt.figure(figsize=(7, 5), dpi=80)
sns.countplot(labels)
plt.title("Count of Labels Assigned",fontsize=12)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.ylabel("count",fontsize=12)
plt.show()
```
